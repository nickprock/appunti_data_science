```python
from datasets import load_dataset

arxiv = load_dataset("effectiveML/ArXiv-10", split="train[4000:5000]")

```

```python
arxiv[0]
```

```python
import re

splitter = re.compile(r'\.\s?\n?')
```

```python
sentences = []
for row in arxiv:
    new_sentences = splitter.split(row['abstract'].replace('\n', ' ').strip())
    new_sentences = [line for line in new_sentences if len(line) > 10]
    sentences.append(new_sentences)
```

```python
from fastembed import TextEmbedding
```

```python
embedding_model = TextEmbedding("BAAI/bge-small-en-v1.5")
print("The model BAAI/bge-small-en-v1.5 is ready to use.")
```

```python
from tqdm import tqdm
```

```python
emb = []
for art in tqdm(sentences):
    emb_chk = [list(embedding_model.embed(chunk, parallel=0)) for chunk in art]
    emb.append(emb_chk)
```

```python
len(emb[0][0][0])
```

```python
from qdrant_client import QdrantClient, models
```

```python
client = QdrantClient(":memory:")
```

```python
points = []
idx=1
doc_id = 1

for i in tqdm(range(len(emb))):
    for j in range(len(emb[i])):
        points.append(models.PointStruct(
            id = idx,
            payload={
                "doc_id": doc_id,
                "title": arxiv[i]['title'].replace('\n', ' ').strip(),
                "text": sentences[i][j],
                "label": arxiv[i]['label']
            },
            vector = emb[i][j][0]
        )
                     )
        idx+=1
    doc_id+=1
```

```python
client.create_collection(
    collection_name="arxiv_6000",
    vectors_config=models.VectorParams(size=384, distance=models.Distance.COSINE),
)
```

```python
client.create_collection(
    collection_name="full_docs",
    vectors_config=models.VectorParams(size=1, distance=models.Distance.COSINE),
)
```

```python
client.upsert(
    collection_name="arxiv_1000",
    points=points
)
```

```python
query = "What is the link between dark matter and supermassive black holes?"

emb_query = list(embedding_model.embed(query, parallel=0))[0]
```

```python
normal_search=client.query_points(
    collection_name="arxiv_1000",
    query=emb_query,
    with_vectors=False,
    with_payload=True,
    limit=10,
).points
```

```python
for hit in normal_search:
    print(hit.payload, "score:", hit.score)
    print("\n")
```

```python
group_search=client.query_points_groups(
    collection_name="arxiv_1000",
    # Same as in the regular query_points() API
    query=emb_query,
    # Grouping parameters
    group_by="doc_id",  # Path of the field to group by
    limit=5,  # Max amount of groups
    group_size=3,  # Max amount of points per group
)
```
